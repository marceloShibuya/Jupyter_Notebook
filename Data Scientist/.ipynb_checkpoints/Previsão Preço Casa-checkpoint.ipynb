# Imports

```python
import pandas as pd
from sklearn import linear_model as lm
import numpy as np
```

# Loading Data

```python
df = pd.read_csv('kc_house_data.csv')
display(df)
```

```python
df.head()
```

```python
df.tail()
```

# Data Preparation

```python
# características
x_train = df.drop(['price','date'], axis = 1)

# o que se deseja de resposta 
y_train = df['price'].copy()
```

# Model Training

```python
# model description
model_lr = lm.LinearRegression()

# model training
model_lr.fit(x_train, y_train)
```

```python
# prediction
pred = model_lr.predict(x_train)
```

```python
pred[0:100]
```

# Performance Metrics

```python
df1 = df.copy()
```

```python
df1['prediction'] = pred
```

```python
df1[['id','price','prediction']]
```

```python
# calculando o erro entre o preço e a predição
df1['error'] = df1['price'] - df1['prediction']
```

```python
df1[['id','price','prediction','error']]
```

```python
# transformando em valores absolutos
df1['error_absolute'] = np.abs(df1['error'])
```

```python
df1[['id','price','prediction','error','error_absolute']]
```

```python
# mae = mean absolute error

mae = np.mean(df1['error_absolute'])
print('MAE: {}'.format(mae))
```

```python
# calcular a porcentagem
df1['error_perc'] = df1['error'] / df1['price']
```

```python
df1[['id','price','prediction','error','error_absolute','error_perc']]
```

```python
# transformando em valores absolutos
df1['error_perc_abs'] = np.abs(df1['error_perc'])
```

```python
df1[['id','price','prediction','error','error_absolute','error_perc','error_perc_abs']]
```

```python
# mean absolute percentage error
mape = np.mean(df1['error_perc_abs'])

print(f'MAPE: {format(mape)}')
# print(f'MAPE: {(mape):.2f}')
```
