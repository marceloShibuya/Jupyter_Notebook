```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
display(df)
```

```python
# Transformar no tipo date

df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

# Criação da coluna: 'house_age'
- Se o valor da coluna 'date' for maior que 2014-01-01 -> 'new_house'
- Se o valor da coluna 'date' for menor que 2014-01-01 -> 'old_house'

```python
df.loc[:,'house_age'] = ''
```

```python
df.loc[df['date'] > '2014-01-01', 'house_age'] = 'new_house'
```

```python
df.loc[df['date'] < '2014-01-01', 'house_age'] = 'old_house'
```

```python
display(df)
```

# Criação da coluna: 'dormitory_type'
- Se o valor da coluna 'bedrooms' for igual a 1 -> 'studio'
- Se o valor da coluna 'bedrooms' for igual a 2 -> 'apartment'
- Se o valor da coluna 'bedrooms' for maior que 2 -> 'house'

```python
df.loc[:,'dormitory_type'] = ''
```

```python
df.loc[df['bedrooms'] == 1, 'dormitory_type'] = 'studio'
```

```python
df.loc[df['bedrooms'] == 2, 'dormitory_type'] = 'apartment'
```

```python
df.loc[df['bedrooms'] > 2, 'dormitory_type'] = 'house'
```

```python
display(df)
```

# Criação da coluna: 'condition_type'
- Se o valor da coluna 'condition' for menor ou igual a 2 -> 'bad'
- Se o valor da coluna 'condition' for igual a 3 ou 4 -> 'regular'
- Se o valor da coluna 'condition' for igual a 5 -> 'good'

```python
df.loc[:,'condition_type'] = ''
```

```python
df.loc[df['condition'] <= 2, 'condition_type'] = 'bad'
```

```python
df.loc[(df['condition'] == 3) | (df['condition'] == 4),'condition_type'] = 'regular'
```

```python
df.loc[df['condition'] == 5, 'condition_type'] = 'good'
```

```python
display(df)
```

# Modificando a coluna 'condition' para STRING

```python
df['condition'] = df['condition'].astype(str)
```

```python
df.info()
```

# Deletando as colunas 'sqft_living15' e 'sqft_lot15'

```python
df = df.drop('sqft_living15', axis = 1)
```

```python
df = df.drop('sqft_lot15', axis = 1)
```

```python
display(df)
```

# Modificando o tipo da coluna 'yr_built' para date

```python
df['yr_built'] = pd.to_datetime(df['yr_built'])
```

# Modificando o tipo da coluna 'yr_renovated' para date

```python
df['yr_renovated'] = pd.to_datetime(df['yr_renovated'])
```

# Quantos imóveis possuem 2 andares

```python
df.loc[df['floors'] == 2,'floors'].count()
```

# Qual o maior número de quartos que um imóvel do tipo house possui

```python
qtd_quartos = df.loc[df['dormitory_type'] == 'house',['dormitory_type','bedrooms']]
qtd_quartos.sort_values('bedrooms',ascending=False)
```

# Qual o valor do imóvel mais caro do tipo studio


```python
imovel_caro = df.loc[df['dormitory_type'] == 'studio',['dormitory_type','price']]
imovel_caro.sort_values('price',ascending=False)
```

# Quantos imóveis estão com a condição good e são new_house


```python
qtd_imoveis = df.loc[(df['condition_type'] == 'good' ) & (df['house_age'] == 'new_house'),['condition_type','house_age']]
qtd_imoveis.value_counts(['condition_type','house_age'])
```

# Quantos imóveis estão com a condição regular


```python
qtd_imoveis_regular = df.loc[df['condition_type'] == 'regular','condition_type']
qtd_imoveis_regular.count()
```
